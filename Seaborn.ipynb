```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
x=[1,2,3,4,5,6,7,8]
y=[9,8,7,4,1,2,5,6]
plt.plot(x,y)
plt.show()
```

```python
data_frame=pd.DataFrame({"x":x,"y":y})
```

```python
sns.lineplot(x="x",y="y",data=data_frame)
plt.show()
```

```python
s_data=sns.load_dataset("penguins")
s_data
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=s_data,hue="sex")
plt.show()
```

```python
sns.displot(s_data["body_mass_g"],
            bins=[2500,3000,3500,4000,4500,5000,5500,6000,6500,7000],kde=True,rug=True,color="y",log_scale=False)
plt.show()
```

```python
sns.set(style="ticks")
sns.barplot(x="species",y="bill_length_mm",data=s_data,hue="sex",saturation=10,errcolor="black",alpha=0.9)
plt.show()
```

```python
m={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=s_data,hue="sex",alpha=1,sizes=(50,200),
               style="sex",palette="bone_r",markers=m)
plt.show()
```

```python
arr=np.array([["a0","a1","a2","a3","a4"],["b0","b1","b2","b3","b4"],["c0","c1","c2","c3","c4"],
             ["d0","d1","d2","d3","d4"]])
ft={"fontsize":20,"color":"w"}
sns.heatmap(var,vmax=10,vmin=0,cmap="twilight_shifted",annot=arr,fmt="s",annot_kws=ft)
plt.show()
```

# Count Plot

```python
cp=sns.load_dataset("tips")
cp
```

```python
sns.countplot(x="sex",data=cp,hue="smoker",palette="gist_heat",saturation=0.8)
plt.show()
```

```python
sns.barplot(x="sex",y="size",data=cp)
plt.show()
```

```python
cp
```

```python
sns.violinplot(x="day",y="total_bill",data=cp,hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=cp,linewidth=2,palette="gist_heat_r")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=cp,hue="sex",palette="tab20c",saturation=1,
              split=True,inner="quart")
plt.show()
```

```python
sns.violinplot(y=cp["size"])
plt.show()
```

```python
sns.pairplot(cp,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="afmhot_r",kind="kde")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=cp,hue="sex",palette="afmhot",linewidth=1,edgecolor="red",)
plt.show()
```

# Box Plot

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=cp,hue="time",order=["Fri","Sun","Sat","Thur"],
           showmeans=True,meanprops={"marker":"+","markeredgecolor":"blue"},linewidth=2,palette="PuBu",
           orient="v")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=cp,showmeans=True,meanprops={"marker":"+","markeredgecolor":"blue"},linewidth=2,palette="PuBu",
           orient="v")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y=cp["total_bill"])
plt.show()
```

# Factor Plot

```python
sns.factorplot(x="size",y="tip",data=cp,hue="sex",kind="box")
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=cp,hue="sex",height=6,kind="point")
plt.show()
```

# Style

```python
sns.set(style="whitegrid")
sns.set_style("darkgrid")
sns.barplot(x="day",y="total_bill",data=cp)
plt.figure(figsize=(10,12))
sns.despine()
plt.show()
```

# Facet Grid

```python
fg=sns.FacetGrid(cp,col="day",hue="sex",palette="RdPu")

fg.map(plt.bar,"total_bill","tip",edgecolor="black").add_legend()
plt.show()
```
